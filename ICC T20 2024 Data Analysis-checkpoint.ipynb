```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Step 1: Fetch the webpage content
url = "https://www.espncricinfo.com/records/tournament/batting-most-runs-career/icc-men-s-t20-world-cup-2024-15946"
response = requests.get(url)
if response.status_code == 200:
    print("Page fetched successfully!")
else:
    print("Failed to fetch the page. Status code:", response.status_code)

# Step 2: Parse the webpage content
soup = BeautifulSoup(response.content, 'html.parser')

# Step 3: Find the table containing the batting records
table = soup.find('table')

if table:
    # Step 4: Extract table headers, if available
    headers = [header.text.strip() for header in table.find_all('th')]
    
    if not headers:  # If headers list is empty, generate generic column names
        num_columns = len(table.find_all('tr')[1].find_all('td'))  # Use the first row for column count
        headers = [f'Column_{i+1}' for i in range(num_columns)]

    # Step 5: Extract table rows
    rows = table.find_all('tr')[1:]  # Skip the header row

    # Step 6: Extract data from rows
    data = []
    for row in rows:
        cols = row.find_all('td')
        cols = [ele.text.strip() for ele in cols]
        data.append(cols)

    # Step 7: Create a DataFrame
    df = pd.DataFrame(data, columns=headers)

    # Step 8: Save the DataFrame to a CSV file
    df.to_csv('icc_t20_world_cup_2024_batting_records.csv', index=False)

    print("Data has been saved to 'icc_t20_world_cup_2024_batting_records.csv'.")
else:
    print("Table not found on the page. Please check the HTML structure or class names.")

print(df.head())
```

```python
import pandas as pd

# Load the batting and bowling datasets
df_batting = pd.read_csv('icc_t20_world_cup_2024_batting_records.csv')
df_bowler = pd.read_csv('icc_t20_world_cup_2024_bowling_records.csv')

# Manually assign the correct column names to both datasets
df_batting.columns = ['Player', 'Span', 'Mat', 'Inns', 'NO', 'Runs', 'HS', 'Ave', 'BF', 'SR', '100', '50', '0', '4s', '6s']
df_bowler.columns = ['Player', 'Span', 'Mat', 'Inns', 'Balls', 'Overs', 'Mdns', 'Runs_Bowling', 'Wkts', 'BBI', 'Ave_Bowling', 'Econ', 'SR_Bowling', '4', '5']

# Extract the required columns from both datasets
df_batting_min = df_batting[['Player', 'Runs', 'SR', 'Mat', 'Inns', '4s', '6s', 'Ave']]
df_bowler_min = df_bowler[['Player', 'Wkts', 'Overs', 'Ave_Bowling', 'Econ']]

# Merge the two datasets on 'Player' column (outer join to keep all players, even if they appear in only one dataset)
df_allrounder = pd.merge(df_batting_min, df_bowler_min, on='Player', how='outer')

# Clean the dataset (convert columns to appropriate types)
df_allrounder['Runs'] = pd.to_numeric(df_allrounder['Runs'], errors='coerce')
df_allrounder['SR'] = pd.to_numeric(df_allrounder['SR'], errors='coerce')
df_allrounder['Ave'] = pd.to_numeric(df_allrounder['Ave'], errors='coerce')
df_allrounder['Wkts'] = pd.to_numeric(df_allrounder['Wkts'], errors='coerce')
df_allrounder['Overs'] = pd.to_numeric(df_allrounder['Overs'], errors='coerce')
df_allrounder['Ave_Bowling'] = pd.to_numeric(df_allrounder['Ave_Bowling'], errors='coerce')
df_allrounder['Econ'] = pd.to_numeric(df_allrounder['Econ'], errors='coerce')
df_allrounder['4s'] = pd.to_numeric(df_allrounder['4s'], errors='coerce')
df_allrounder['6s'] = pd.to_numeric(df_allrounder['6s'], errors='coerce')

# Display the first few rows of the all-rounder dataset
print(df_allrounder.head(100))

# Save the combined all-rounder dataset to a CSV
df_allrounder.to_csv('icc_t20_world_cup_2024_allrounder_minimal_records.csv', index=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from math import pi

# Set style
sns.set(style="whitegrid")

# 1. Batting Performance (Top 4 Batsmen)
plt.figure(figsize=(10, 6))
top_4_batsmen = best_11_batsmen
sns.barplot(x='Player', y='Runs', data=top_4_batsmen, palette='viridis')
plt.title('Top 4 Batsmen by Runs')
plt.xlabel('Player')
plt.ylabel('Runs')
plt.xticks(rotation=45)
plt.show()

# 2. Bowling Performance (Top 4 Bowlers)
plt.figure(figsize=(10, 6))
top_4_bowlers = best_11_bowlers
sns.barplot(x='Player', y='Wkts', data=top_4_bowlers, palette='Blues_d')
plt.title('Top 4 Bowlers by Wickets')
plt.xlabel('Player')
plt.ylabel('Wickets')
plt.xticks(rotation=45)
plt.show()

# 3. All-rounder Radar Chart (Top 3 All-rounders)
# Prepare data for radar chart
top_3_allrounders = best_11_allrounders

categories = ['Runs', 'SR', 'Mat', 'Overs', 'Ave', 'Econ', '4s', '6s']
values = top_3_allrounders[categories].values.T
labels = top_3_allrounders['Player'].values

# Number of categories
N = len(categories)

# Angle for each axis
angles = [n / float(N) * 2 * pi for n in range(N)]
angles += angles[:1]

# Plot radar charts for each all-rounder
for i, player in enumerate(labels):
    values_for_player = values[:, i].tolist()
    values_for_player += values_for_player[:1]
    
    fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
    ax.set_theta_offset(pi / 2)
    ax.set_theta_direction(-1)
    ax.set_rlabel_position(0)
    
    ax.plot(angles, values_for_player, linewidth=1, linestyle='solid', label=player)
    ax.fill(angles, values_for_player, alpha=0.4)
    
    ax.set_yticklabels([])
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(categories, rotation=45)
    
    plt.title(f'All-rounder Performance for {player}')
    plt.legend(loc='upper right', bbox_to_anchor=(1.1, 1.1))
    plt.show()
```
